```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from scipy.interpolate import griddata
```

```python
# Meshgrid
x, y = np.meshgrid(np.linspace(-5, 5, 10),
				np.linspace(-5, 5, 10))

# Defining argument combinations
scaled_kwargs = {'scale_units':'xy', 'scale':1, 'angles':'xy'}
default_kwargs = {'scale_units':None, 'scale':None, 'angles':'uv'}

# Inputs are the direction at each x, y point
def plot_vector_field(u, v, kwargs=default_kwargs):
    # Creating figure
    plt.figure(figsize=(8,8))

    # Plotting Vector Field with QUIVER (x, y are positions, u, v are directions)
    plt.quiver(x, y, u, v, color='black', **kwargs)
    plt.scatter(x,y, color='blue', marker='x')
    plt.title('Vector Field')
    plt.xlabel('x')
    plt.ylabel('y')

    # Setting x, y boundary limits
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)

    # Show plot with grid
    plt.grid()
    plt.show()
```

```python
# Vector (0.5, 1)
u = 0.5 
v = 1
plot_vector_field(u,v, scaled_kwargs)
```

```python
u = x/(np.sqrt(x**2 + y**2))
v = y/(np.sqrt(x**2 + y**2))
plot_vector_field(u,v)
```

```python
u = x/(np.sqrt(x**2 + y**2)**2)
v = y/(np.sqrt(x**2 + y**2)**2)
plot_vector_field(u,v)
```

```python
u = -y
v = x
plot_vector_field(u,v)
```

```python
u = y
v = x
plot_vector_field(u,v)
```

```python
u = -np.sqrt(x**2 + y**2)*y
v = np.sqrt(x**2 + y**2)*x
plot_vector_field(u,v)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')

x3, y3, z3 = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

u = np.sin(np.pi * x3) * np.cos(np.pi * y3) * np.cos(np.pi * z3)
v = -np.cos(np.pi * x3) * np.sin(np.pi * y3) * np.cos(np.pi * z3)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x3) * np.cos(np.pi * y3) *
     np.sin(np.pi * z3))

ax.quiver(x3, y3, z3, u, v, w, length=0.1, color = 'black')

plt.show()
```
